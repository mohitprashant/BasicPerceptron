```python
import matplotlib.pyplot as plt
import random
```

```python
#Target classifier line, ax+by+c=0

a=5
b=1
c=0
```

```python
#Generate points

n=100

x1=[]
y1=[]

x2=[]
y2=[]

for i in range(n):
    x=random.randint(-100, 100)
    y=random.randint(-100, 100)
    
    if(a*x+b*y+c<0):
        x1.append(x)
        y1.append(y)
    else:
        x2.append(x)
        y2.append(y)
```

```python
#Show plot

fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])

ax.scatter(x=x1, y=y1, marker='o', color='b')
ax.scatter(x=x2, y=y2, marker='^', color='r')

plt.xlim(-100, 100)
plt.ylim(-100, 100)

plt.show()
```

```python
#Learning process


#Prediction function
def predict(w, x, y):
    if(w[0]*x+w[1]*y+w[2]<0):
        return 0
    else:
        return 1

    
#Update weights vector, data in format ---> [[x1, y1], [x2, y2]]
def updateWeights(w, data):
    r1=random.randint(0, 1)
    r2=random.randint(0, len(data[r1][0])-1)
    
    x=data[r1][0][r2]
    y=data[r1][1][r2]
    
    
    if(predict(w, x, y)==r1):
        return 0
    elif(r1==0):
        w[0]-=x
        w[1]-=y
        return 1
    elif(r1==1):
        w[0]+=x
        w[1]+=y
        return 1
    

#Show the results of the prediction, data in format ---> [[x1, y1], [x2, y2]] 
def showPredict(w, data):
    fig=plt.figure()
    ax=fig.add_axes([0, 0, 1, 1])
    
    ax.scatter(x=data[0][0], y=data[0][1], marker='o', color='b')
    ax.scatter(x=data[1][0], y=data[1][1], marker='^', color='r')
    
    xp=[]
    yp=[]
    
    for i in range(-100, 101):
        xp.append(i)
        yp.append(-(w[0]/w[1])*i-(w[2]/w[1]))
        
    ax.scatter(x=xp, y=yp, marker='.')
    
    
    plt.xlim(-100, 100)
    plt.ylim(-100, 100)

    plt.show()
```

```python
#Initialize weights
w=[random.randint(-10, 10), random.randint(-10, 10), 1]
showPredict(w, [[x1, y1], [x2, y2]])
w
```

```python
for i in range(10):
    updateWeights(w, [[x1, y1], [x2, y2]])
    showPredict(w, [[x1, y1], [x2, y2]])
    print(w)
```
